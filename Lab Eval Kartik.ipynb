```python
student = {}

def  add_student():
    name = input("Enter the name of the student")
    batch = input("Enter the batch of the student")
    maths = input("Enter the grade of maths")
    phy = input("Enter the grade of phy")
    chem = input("Enter the grade of chemistry")
    student[name] = {
        name: name,
        "batch": batch,
        "maths": int(maths),
        "phy": int(phy),
        "chem": int(chem),
    };

def update_grades():
    name = input("Enter the name of the student")
    if(name in student):
        print("Enter 1. maths")
        print("Enter 2. physics")
        print("Enter 3. chemistry")
        print("Enter 4. exit")
        while(1):
            choice = input("Enter the choice of subject")
            if(choice == "1"):
                maths = input("Enter the grade of maths")
                student[name]["maths"] = int(maths)
                
            elif(choice == "2"):
                phy = input("Enter the grade of phy")
                student[name]["phy"] = int(phy)
                
            elif(choice == "3"):   
                chem = input("Enter the grade of chemistry")
                student[name]["chem"] = int(chem)
                
            elif(choice == "4"):  
                break;
                
            else: 
                print("enter valid choiuce")
    else:
        print("No Student Record")
        
def calculate_average():
    name = input("Enter the name of the student")
    if(name in student):
        details = student[name]
        tsum = details["maths"] + details["chem"] + details["phy"]
        print("Average is ", tsum/3)
    else:
        print("No Student Record")      
        
def generate_top_students_report():
    top_student = {}
    maxi = 0
    for name, details in student.items():
        tsum = details["maths"] + details["chem"] + details["phy"]
        avg = tsum/3
        if avg > maxi:
            maxi = avg
            top_student.clear()
            top_student[name] = details
            
    name = top_student.keys()[0]
    print("Report of top student is ")
    print(f"Student Name {top_student.keys[0]} ")
#     print(f"Maths Grade is {top_student[name]["maths"]}")
#     print(f"Physics Grade is {top_student[name]["phy"]}")
#     print(f"Chemistry Grade is {top_student[name]["chem"]}")
```

```python
print("Enter 1. For Adding Student")
print("Enter 2. For Updating Grades of Student")
print("Enter 3. For Calculating  Student")
print("Enter 4. For Adding Student")
print("Enter 5. For exit")

while(1):
    choice = input("Enter the choice")
    
    if(choice == "1"):
        add_student()
    elif(choice == "2"):
        update_grades()
    elif(choice == "3"):
        calculate_average()
    elif(choice == "4"):
        generate_top_students_report()
    elif(choice == "5"):
        break;
    else:
        print("Enter valid Choice")
    
```
